# 2. Conceitos Importantes

- <font color='darkcian'>**MÉDIA:**</font> soma de todos os elementos dividido pelo número de elementos n

- Comando: __np.mean(variável)__

- <font color='darkcian'>**VARIÂNCIA:**</font> mede dispersão dos dados
- Comando: __np.var(variável)__

- <font color='darkcian'>**MEDIANA:** </font>Corta os dados no meio, não é afetada por medidas extremas (como a média)
- Comando: __np.median(variável)__

- <font color='darkcian'>**DESVIO PADRÃO:**</font> é uma espécie de desvio médio dos valores em relação à média aritmética
- Comando: __np.std(variável)__

- <font color='darkcian'>**MODA: </font>**valor com maior frequência nos dados
- Comando: __stats.mode(variável)__

____

- <font color='darkcian'>**VARIÁVEL ALEATÓRIA:**</font> função que associa um número real a cada ponto do espaço amostral (possível realização do experimento aleatório)

- <font color='darkcian'>**VALOR ESPERADO:**</font> somados produtos dos valores da variável pelas respectivas probabilidades da variável assumir tais valores

## COVARIÂNCIA E CORRELAÇÃO
- __Medidas do grau de interdependência (ou inter-relação) numérica entre duas variáveis aleatórias__
- __A correlação mede tanto a força como a direção da relação linear entre duas variáveis. Um relacionamento linear perfeito resulta em um coeficiente de correlação 1.__
- __O coeficiente de correlação é uma função da covariância. O coeficiente de correlação é igual à covariância dividida pelo produto dos desvios padrão das variáveis.__

__Comandos:__

</font>

p = probabilidade de sucesso

E[X] = np

<font color='purple'>
    
**pmf**: Variáveis aleatórias discretas </font>

$P(Y=y)$: prob = stats.binom.pmf(y, n, p)__

probablidade de y sucessos em n tentaivas com probabilidade p

<font color='purple'>**cdf**: Variáveis aleatórias cumulativas </font>

$P(Y\leq y)$: prob = stats.binom.cdf(y, n, p)

__$E(Y)$ e $Var(Y)$: m, v = stats.binom.stats(n, p, moments='mv')__

#### Os β e os P

P alto: não rejeito H0

P baixo: rejeito H0 (β não presta, não tem impacto)

#### Impacto do β

Impacto pra cima: β positivo

Impacto pra baixo: β negativo

#### P > |t| & F-statistic

Probabilidade de que as constantes sejam 0

P > |t|: para cada coeficiente

F-statistic: probabilidade de que todos sejam 

H0: todos os β exceto pelo constante sõa nulos

Rejeito: pelo menos algum β não é nulo

Se é < alpha, meu modelo presta para alguma coisa

#### Omnibus e Jarque_Bera (JB)

H0: resíduos seguem distribuição normal

p do JB baixo: rejeito a hipótese nula - mesma coisa pro Omnibus

#### Durbin-Watson

Próximo de 2: ruídos não são relacionados

1 ou 4: ruídos relacionados

Mais importante para sinais temporais (se 2 instantes de tempo se relacionam)
